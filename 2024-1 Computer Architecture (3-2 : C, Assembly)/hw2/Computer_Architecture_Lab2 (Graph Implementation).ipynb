```python
import pandas as pd
import matplotlib.pyplot as plt
```

### trace-1.output

```python
# 텍스트 파일에서 데이터 읽어오기
file_path = '/content/trace1-out.txt'  # 텍스트 파일 경로 설정
with open(file_path, 'r') as file:
    lines = file.readlines()
```

```python
# 데이터 처리를 위한 빈 리스트 생성
data = {
    'policy': [],
    'cache_size': [],
    'block_size': [],
    'associative': [],
    'd-miss_rate': [],
    'i-miss_rate': [],
    'mem_write': []
}

# 텍스트 파일 데이터를 데이터프레임에 넣기
for line in lines:
    line = line.strip()
    if line.startswith('LRU') or line.startswith('Random'):
        parts = line.split()
        data['policy'].append(parts[0])
        data['cache_size'].append(int(parts[1]))
        data['block_size'].append(int(parts[2]))
        data['associative'].append(int(parts[3]))
        data['d-miss_rate'].append(float(parts[4]))
        data['i-miss_rate'].append(float(parts[5]))
        data['mem_write'].append(int(parts[6]))

df = pd.DataFrame(data)

# LRU와 Random에 대해 각각 그래프 그리기
associative_values = [1, 2, 4, 8]
block_sizes = [16, 64]
```

```python
import numpy as np

# Set a larger figure size
plt.figure(figsize=(12, 10))

# Define associative values as strings
associative_labels = {1: '1-way', 2: '2-way', 4: '4-way', 8: '8-way'}

# Set common y-axis limits for all plots
y_min = 0
y_max = 0.3

# LRU/16 (I-Cache) 그래프 그리기
plt.subplot(2, 2, 1)
for associative in associative_values:
    df_plot = df[(df['policy'] == 'LRU') & (df['block_size'] == 16) & (df['associative'] == associative)]
    plt.plot(df_plot['cache_size'], df_plot['i-miss_rate'], label=f'LRU/{associative_labels[associative]} (I-Cache), 16 Bytes')
plt.title('LRU / 16 (I-cache)')
plt.xlabel('Cache Size')
plt.ylabel('Indirect Miss Rate')
plt.ylim(y_min, y_max)  # Set y-axis limits
plt.yticks(np.arange(y_min, y_max + 0.05, 0.05))  # Set y-axis ticks in increments of 0.05
plt.legend()

# LRU/64 (I-Cache) 그래프 그리기
plt.subplot(2, 2, 2)
for associative in associative_values:
    df_plot = df[(df['policy'] == 'LRU') & (df['block_size'] == 64) & (df['associative'] == associative)]
    plt.plot(df_plot['cache_size'], df_plot['i-miss_rate'], label=f'LRU/{associative_labels[associative]} (I-Cache), 64 Bytes')
plt.title('LRU / 64 (I-cache)')
plt.xlabel('Cache Size')
plt.ylabel('Indirect Miss Rate')
plt.ylim(y_min, y_max)  # Set y-axis limits
plt.yticks(np.arange(y_min, y_max + 0.05, 0.05))  # Set y-axis ticks in increments of 0.05
plt.legend()

# LRU/16 (D-Cache) 그래프 그리기
plt.subplot(2, 2, 3)
for associative in associative_values:
    df_plot = df[(df['policy'] == 'LRU') & (df['block_size'] == 16) & (df['associative'] == associative)]
    plt.plot(df_plot['cache_size'], df_plot['d-miss_rate'], label=f'LRU/{associative_labels[associative]} (D-Cache), 16 Bytes')
plt.title('LRU / 16 (D-cache)')
plt.xlabel('Cache Size')
plt.ylabel('Direct Miss Rate')
plt.ylim(y_min, y_max)  # Set y-axis limits
plt.yticks(np.arange(y_min, y_max + 0.05, 0.05))  # Set y-axis ticks in increments of 0.05
plt.legend()

# LRU/64 (D-Cache) 그래프 그리기
plt.subplot(2, 2, 4)
for associative in associative_values:
    df_plot = df[(df['policy'] == 'LRU') & (df['block_size'] == 64) & (df['associative'] == associative)]
    plt.plot(df_plot['cache_size'], df_plot['d-miss_rate'], label=f'LRU/{associative_labels[associative]} (D-Cache), 64 Bytes')
plt.title('LRU / 64 (D-cache)')
plt.xlabel('Cache Size')
plt.ylabel('Direct Miss Rate')
plt.ylim(y_min, y_max)  # Set y-axis limits
plt.yticks(np.arange(y_min, y_max + 0.05, 0.05))  # Set y-axis ticks in increments of 0.05
plt.legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
