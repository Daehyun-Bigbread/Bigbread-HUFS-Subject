## Problem 1 (5 pts)

- The centroid of a finite set of $\displaystyle {k}$ points $\displaystyle \mathbf {x} _{1},\mathbf {x} _{2},\ldots ,\mathbf {x} _{k}$ in $\displaystyle \mathbb {R} ^{n}$ is
$$
\mathbf {C} ={\frac {\mathbf {x} _{1}+\mathbf {x} _{2}+\cdots +\mathbf {x} _{k}}{k}}
$$

- This point minimizes the sum of squared Euclidean distances between itself and each point in the set.
- Compute centroid
- Plot dataset and centroid

---

- $\displaystyle {k}$ 포인트의 유한 집합 $\displaystyle \mathbf {x} _{1},\mathbf {x} _{2},\ldots,\mathbf {x} _{k}$의 중심은
$$
\mathbf {C} ={\frac {\mathbf {x} _{1}+\mathbf {x} _{2}+\cdots +\mathbf {x} _{k}}{k}}
$$

- 이 점은 자신과 집합의 각 점 사이의 유클리드 거리 제곱의 합을 최소화합니다.
- 계산 중심
- 데이터 집합 및 중심 그림

````python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

def plot_centroid(data):
  centroid = np.mean(data, axis=0)
  plt.scatter(data[:,0], data[:,1]) # x 좌표[0], y 좌표[1]
  plt.scatter(centroid[0], centroid[1], marker='s', color='red')
````

## Problem 2 (10 pts)

- Let $\textstyle x_{1},x_{2},\cdots ,x_{n}$ be a set of $\textstyle n$ points in a space with a distance function $d$.
- Medoid is defined as
$$
x_{\text{medoid}} = \text{argmin}_{y \in \{x_1, x_2, \cdots, x_n\}} \sum_{i=1}^{n}d(y, x_i)
$$
- Compute medoid using Euclidean distance as a distance function.
- Plot dataset and medoid
- *Do not use sklearn, scipy or any module computing distance matrix directly*
- Use numpy functions only

## 문제 2(10pts)

- 거리 함수 $d$가 있는 공간에서 $\textstyle x_{1}, x_{2},\cdots,x_{n}$를 $\textstyle n$ 포인트 집합이라고 합니다.
- Medoid는 다음과 같이 정의됩니다
$$
x_{\text{medoid}} = \text{argmin}_{y \in \{x_1, x_2, \cdots, x_n\}} \sum_{i=1}^{n}d(y, x_i)
$$
- 유클리드 거리를 거리 함수로 사용하여 메도이드를 계산합니다.
- 데이터 세트 및 메도이드 그림
- *sklearn, scippy 또는 모듈 컴퓨팅 거리 매트릭스를 직접 사용하지 마십시오*
- numpy 함수만 사용

````python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

def plot_medoid(data):

# YOUR CODE MUST BE HERE
# median의 거리, -> medoid. y - 이중에 한개

  # 포인트 간의 거리 행렬 계산
  distance = np.sqrt(np.sum((data[:,np.newaxis,:] - data[np.newaxis,:,:])**2, axis=2))

  # 거리 행렬에서 각 데이터 포인트와의 거리의 합이 최소가 되는 Medoid의 인덱스 찾기
  medoid = np.argmin(np.sum(distance, axis=1))

  # 데이터 포인트를 산점도로 그리기
  plt.scatter(data[:,0], data[:,1]) # x 좌표[0], y 좌표[1]

  # Medoid를 빨간색 사각형으로 표시
  plt.scatter(data[medoid,0], data[medoid,1], marker='s', color='red')

  # 그래프 표시
  plt.show()
````

````python
# DO NOT EDIT THIS CELL
# RUN THIS CELL

np.random.seed(0)
data = np.random.randn(10,2)
plot_medoid(data)
````

## Problem 3 (5 pts)

- We want to draw a scatter plot using **data**
- Plot the center using a green square symbol
- Plot points inside **radius** from center using red dots
- Plot points out of the **radius** from center using 'C0' colored dots
- Draw a filled circle centered at **center** using red color and alpha=0.2

- **data**를 사용하여 산점도를 그리고자 합니다
- 녹색 사각형 기호를 사용하여 중심 표시
- 빨간색 점을 사용하여 중심에서 **반지름** 내부에 점 그리기

- 빨간색과 알파=0.2를 사용하여 **center**를 중심으로 채워진 원을 그립니다

````python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

def points_within_radius(data, center, radius):
  # YOUR CODE MUST BE HERE
  # 중심과 데이터 포인트 사이의 거리 계산
  distances = np.sqrt(np.sum((data - center)**2, axis=1))

  # 중심과 반지름 내부에 있는 점 선택
  inner_points = data[distances < radius]

  # 산점도 그리기
  plt.scatter(data[:, 0], data[:, 1], color='C0')

  # 중심 표시
  plt.scatter(center[0], center[1], marker='s', color='green')

  # 중심에서 반지름 내부에 있는 점 그리기
  plt.scatter(inner_points[:, 0], inner_points[:, 1], color='red')

  # 중심으로부터 반지름을 나타내는 원 그리기
  radius_circle = plt.Circle(center, radius, color='red', alpha=0.2)
  plt.gca().add_patch(radius_circle)
  plt.axis('equal')  # 동일한 비율로 축 조정
  plt.show()
````

## Problem 4 (10 pts)

- We want to find k nearest points from the center
- Plot the center using a green square symbol
- Plot k-nearest points from center using red dots
- Plot other points using 'C0' colored dots
- Draw a filled circle centered at **center** using red color and alpha=0.2
- *Do not use sklearn, scipy or any module computing k-nearest points directly*
- Use numpy functions only

- 우리는 중심에서 가장 가까운 k개의 지점을 찾고 싶습니다
- 녹색 사각형 기호를 사용하여 중심 표시
- 빨간색 점을 사용하여 중심에서 가장 가까운 k-점 표시
- 'C0' 색 점을 사용하여 다른 점 표시
- 빨간색과 알파=0.2를 사용하여 **center**를 중심으로 채워진 원을 그립니다
- *klearn, scippy 또는 k-근접 지점을 직접 계산하는 모듈을 사용하지 마십시오*
- numpy 함수만 사용

````python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

def points_k_nearest(data, center, k=1):
    # 중심으로부터 각 데이터 포인트까지의 거리 계산
    distances = np.sqrt(np.sum((data - center)**2, axis=1))

    # 중심으로부터 가장 가까운 k개의 점 찾기
    k_nearest_indices = np.partition(distances, k)[k-1]

    # 중심에서 반지름 내부에 있는 점 선택
    inner_points = data[distances <= k_nearest_indices]

    # 산점도 그리기
    plt.scatter(data[:, 0], data[:, 1], color='C0', label='Other Points')  # 다른 점 표시
    plt.scatter(center[0], center[1], marker='s', c='green', label='Center')  # 중심 표시
    plt.scatter(inner_points[:, 0], inner_points[:, 1], color='red', label='Closest Points')  # 중심으로부터 가장 가까운 k개의 점 표시

    # 중심으로부터 반지름을 나타내는 원 그리기
    radius = np.sort(distances)[k-1]
    radius_circle = plt.Circle(center, radius, color='red', alpha=0.2, fill=True)
    plt.gca().add_patch(radius_circle)
    plt.axis('equal')  # 동일한 비율로 축 조정
    plt.show()
````

````python
# DO NOT EDIT THIS CELL
# RUN THIS CELL

np.random.seed(1)
data = np.random.randn(10,2)
k = 5
center = np.random.randn(2)
points_k_nearest(data, center, k)
````

- **find_k_nearest_index**는 k-nearest의 인덱스를 반환합니다
- 우리는 실행 시간을 맞추기를 원합니다
- *klearn, scippy 또는 k-근접 지점을 직접 계산하는 모듈을 사용하지 마십시오*
- numpy 함수만 사용
- **코드의 numpy 데이터 유형을 변경하지 마십시오**

````python
import numpy as np

# YOUR CODE MUST BE HERE

def find_k_nearest_index(data, center, k=1):
    distance = np.sqrt(np.sum((data - center) ** 2, axis=1))
    k_nearest_indices = np.partition(distance, k-1)[k-1]
    result = np.where(distance <= k_nearest_indices)[0]
    return result
````

## Problem 6 (30 pts)

- **find_k_nearest_index_big** returns the index of the k-nearest for 50, 70, or 100 million data
- We want to time the execution
- *Do not use sklearn, scipy or any module computing k-nearest points directly*
- Use numpy functions only
- You may or have to restart your kernel(runtime) to claim more main memory

- **find_k_nearest_index_big**는 50, 70 또는 1억 데이터에 대한 k-nearest의 인덱스를 반환합니다
- 우리는 실행 시간을 맞추기를 원합니다
- *klearn, scippy 또는 k-근접 지점을 직접 계산하는 모듈을 사용하지 마십시오*
- numpy 함수만 사용
- 더 많은 메인 메모리를 요청하기 위해 커널(실행 시간)을 다시 시작하거나 다시 시작해야 할 수 있습니다
- **코드의 numpy 데이터 유형을 변경하지 마십시오**

````python
import numpy as np

# YOUR CODE MUST BE HERE

def find_k_nearest_index_big(data, center, num_neighbors=1):
    memory_limit = 2 * 1024 * 1024 * 1024  # 2GB 메모리 한도
    each_data_size = data.itemsize * data.shape[1]
    segment_size = memory_limit // (each_data_size * 10)  # 메모리 한도를 고려한 세그먼트 크기 계산

    closest_dist = np.full(num_neighbors, np.inf)  # 가장 가까운 거리를 저장할 배열, 무한으로 초기화
    closest_idx = np.zeros(num_neighbors, dtype=int)  # 가장 가까운 인덱스를 저장할 배열 초기화

    for seg_start in range(0, data.shape[0], segment_size):
        seg_end = min(seg_start + segment_size, data.shape[0])
        segment = data[seg_start:seg_end]

        # 세그먼트와 피벗 포인트 간의 거리 계산
        dist = np.linalg.norm(segment - center, axis=1)
        seg_indices = np.arange(seg_start, seg_end)

        # 현재 세그먼트의 거리 및 인덱스와 기존의 가장 가까운 거리 및 인덱스 결합
        all_dist = np.concatenate((closest_dist, dist))
        all_indices = np.concatenate((closest_idx, seg_indices))

        # 결합된 거리에서 가장 가까운 num_neighbors개 선택
        selected_indices = np.argpartition(all_dist, num_neighbors)[:num_neighbors]
        closest_dist = all_dist[selected_indices]
        closest_idx = all_indices[selected_indices]

    # 최종적으로 찾아진 num_neighbors개의 인덱스를 거리에 따라 정렬
    final_sorted_indices = closest_idx[np.argsort(closest_dist)]

    return final_sorted_indices
````

````python
!uname -a
# DO NOT EDIT THIS CELL
# RUN THIS CELL

import time

np.random.seed(0)
data = np.random.randn(50000000,20)   # 50 million data
k = 5
center = np.random.randn(20)
start = time.time()
print(find_k_nearest_index_big(data, center, k))
end = time.time()

lapse = end - start
total = 10
weight = 1.5
grace = 15
my_point = int(total / (weight ** (lapse // grace)))
print(f'Total time taken : {lapse} seconds')
print(f'My point is {my_point}')
````
